#4. Cleaning the data

##4.1 Removing Outliers

1. **Calculate Q1 and Q3:**

The function calculates the **first quartile (Q1)** and the** third quartile (Q3)** for each column.


2. **Calculate IQR:**

**IQR is the range between Q1 and Q3,** representing the middle 50% of the data. This helps in identifying extreme values.

**IQR = Q3 - Q1.**

3. **Define the Bounds for Outliers:**

Lower Bound: Q1 - 1.5 * IQR.

Upper Bound: Q3 + 1.5 * IQR.

**Any data point outside these bounds is considered an outlier.**

4. **Remove Outliers:**

The function filters the dataset, keeping only the rows where the values in the specified columns are within the lower and upper bounds.

5. **Return the Cleaned Dataset:**

The filtered dataset is returned, and you check the shape of the dataset before and after removing outliers.

```python
# Define a function to identify and remove outliers
def remove_outliers(data, columns):
    for col in columns:
        # Calculate the first quartile (Q1) and third quartile (Q3)
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)

        # Calculate the Interquartile Range (IQR)
        IQR = Q3 - Q1

        # Define the bounds for outliers
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Filter out outliers
        data = data[(data[col] >= lower_bound) & (data[col] <= upper_bound)]

    return data

# List of columns to check for outliers
columns_to_check = [
    'fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
    'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
    'pH', 'sulphates', 'alcohol'
]

# Remove outliers from the specified columns
data = remove_outliers(data, columns_to_check)

# Check the shape of the dataframe before and after removing outliers
print(f"Original dataset shape: {data.shape}")
print(f"Cleaned dataset shape: {data.shape}")
```

#6. Model Building and Evaluation

##6.1 Create Binary Target Variable
This code creates a new column in the dataset, **quality_binary**, where:

**Good wines (with a quality rating of 7 or higher) are labeled as 1.**

**Not Good wines (with a quality rating below 7) are labeled as 0.**

This conversion simplifies the prediction task to a **binary classification** problem (**good vs. not good**).

(**data[target] >= 7**): This **returns True** for wines with a quality rating of** 7 or more,** and **False for those below**.
.astype(int): **Converts True to 1 (good)** and **False to 0 (not good)**

##Features and Target
**X:** This stores the features (**input variables**) that will be used to predict wine quality. It's a subset of the dataset containing only the columns listed in the features list (i.e., alcohol, sulphates, etc.).

**y:** This stores the target variable (**quality_binary**), which the model will learn to predict (**either 1 for good quality or 0 for not good quality**).

```python
# Define the feature columns and target variable
features = ['alcohol', 'sulphates', 'citric acid', 'volatile acidity', 'total sulfur dioxide']
target = 'quality'

# Create binary target variable (Good: 1, Not Good: 0)
data['quality_binary'] = (data[target] >= 7).astype(int)

# Features and target
X = data[features]
y = data['quality_binary']
```

##6.2 Training and Testing

**train_test_split:** This function splits the dataset into two parts:

**70% of the data goes into the training set** (X_train, y_train), which the model will use to learn patterns.

**30% goes into the testing set** (X_test, y_test), which will be used to evaluate how well the model performs on unseen data.

**test_size=0.3:** Specifies that 30% of the data should be reserved for testing, while 70% is for training.

**random_state=42:** Ensures the split is reproducible — running the code multiple times will always result in the same split.

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

##6.3 Build and Train Model Random Forest

###Random Forest:

```python
# Initialize and train the Random Forest model

#Builds multiple decision trees during training and merges their outputs to make predictions
#This ensures that the random processes within the algorithm
#(e.g., selecting random subsets of the data) are consistent each time you
#run the code, providing reproducibility.
random_forest = RandomForestClassifier(random_state=42)

#Trains the Random Forest model on the training data (X_train features, and y_train labels).
#It learns the relationship between the input features and the target
#(wine quality) by constructing multiple decision trees.
random_forest.fit(X_train, y_train)

# Predict on the test set
y_pred_forest = random_forest.predict(X_test)

# Evaluate the Random Forest model
print("Random Forest Performance:")
print(confusion_matrix(y_test, y_pred_forest))
print(classification_report(y_test, y_pred_forest))
```
